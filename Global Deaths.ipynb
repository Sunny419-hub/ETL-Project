# Daily Death Cases

```python
import pandas as pd
import csv
import os
import numpy as np
from sqlalchemy import create_engine
from config import access
```

### Store CSV into DataFrame

```python
csv_path="resources/RAW_global_deaths.csv"
```

```python
deaths_df = pd.read_csv(csv_path)
```

```python
deaths_df.head()
```

```python
deaths_df.replace(np.nan, '', regex=True, inplace=True)
```

```python
deaths_df.loc[deaths_df['Province/State']!='']
```

### Create a summary variable

```python
deaths_df['Total Death Cases']= deaths_df.iloc[:, 320:497].sum(axis=1)  
deaths_df
```

```python
clean_deaths_df = deaths_df[['Country/Region', 'Lat','Long','Total Death Cases']].copy()
```

```python
clean_deaths_df
```

```python
clean_deaths_df.loc[clean_deaths_df['Country/Region']=='Australia']
```

### Keep the required variables 

```python
clean_deaths_df =  clean_deaths_df.groupby(["Country/Region"]).sum()
clean_deaths_df.reset_index(inplace=True)
clean_deaths_df.columns = ['country','total_deaths']
clean_deaths_df
```

### Connect to local database

```python
# Create Engine for SQL Data
engine = create_engine(f'postgresql://postgres:{access}@localhost/etl_db')
conn= engine.connect()
```

```python
engine.table_names()
```

```python
clean_deaths_df.to_sql(name='deaths', con=engine, if_exists='append', index=False)
```
