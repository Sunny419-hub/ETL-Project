```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import csv
import os
```

### Store CSV into DataFrame


```python
csv_file = "../ETL-Project/Resources/RAW_global_confirmed_cases.csv"
confirmed_cases_df = pd.read_csv(csv_file)
confirmed_cases_df.head()
```

### Create a summary variable

```python
 #new variable with total confirmed cases until may 28th 2021
confirmed_cases_df['total_confirmed']= confirmed_cases_df.iloc[:, 320:497].sum(axis=1)  
confirmed_cases_df
```

```python
confirmed_cases_df= confirmed_cases_df.rename(columns={'Country/Region': 'country','Lat': 'lat','Long': 'long'})
confirmed_cases_df
```

### Keep the required variables 

```python
confirmed_cases_df=confirmed_cases_df.loc[:, ["country", "lat", "long", "total_confirmed"]]
confirmed_cases_df
```

### Connect to local database


```python
# Create Engine for SQL Data
engine = create_engine('postgresql://postgres:sql2021@localhost/etl_db')
conn= engine.connect()
```

```python
engine.table_names()
```

```python
confirmed_cases_df.to_sql(name='confirmed_cases', con=engine, if_exists='append', index=False)
```
