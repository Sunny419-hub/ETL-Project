```python
# Dependencies

import csv
import pandas as pd
import os
```

```python
csvpath = ("Resources/worldometer_coronavirus_daily_data.csv")
```

```python
coronavirus_daily = pd.read_csv(csvpath)
```

```python
coronavirus_daily
```

```python
corona_date = coronavirus_daily.loc[(coronavirus_daily["date"] >= "2020-12-2") & (coronavirus_daily["date"] <= "2021-5-28")]
```

```python
corona_date
```

```python
corona_date_country = corona_date.groupby(["country"]).sum()
```

```python
corona_date_country
```

```python
corona_date_country.columns = ["Cumulative Total Cases","Daily New Cases","Active Cases",
                               "Cumulative Total Deaths", "Daily New Deaths"]
```
